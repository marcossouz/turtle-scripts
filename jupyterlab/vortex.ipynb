```python
import turtle
pen = turtle.Turtle()
```

```python
def circle(r):
    pen.pensize(1.5)
    pen.pu()
    pen.rt(90)
    pen.fd(r)
    pen.lt(90)
    pen.pd()
    pen.circle(r)
```

```python
def center():
    pen.pu()
    pen.setpos(0.0, 0.0)
    pen.pd()
```

```python
def marcacao(r):
    pen.lt(50)
    for i in range(9):
        center()
        pen.pu()
        pen.fd(r+25)
        pen.write(f"{i+1}", font=("Arial", 16, "normal"))
        pen.rt(40)
    pen.rt(50)
    center()
```

```python
def pontos(r):
    pen.shape('circle')
    pen.shapesize(0.3, 0.3, 0.3)
    pen.lt(50)
    for i in range(9):
        center()
        pen.pu()
        pen.fd(r)
        pen.stamp()
        pen.rt(40)
    pen.rt(50)
    center()
    pen.shape('classic')
    pen.shapesize(1, 1, 1)
```
